```python
import pandas as pd
import json
import math

# Function to process the excel file and generate the JSON structure
def excel_to_json(file_path, sheet_names, pillar_names):
    missions = {}
    missions_with_rec_ids = set()

    # Read the excel file
    excel_data = pd.ExcelFile(file_path)

    recommendations = {}

    # Process each sheet
    for i, sheet in enumerate(sheet_names):
        # Read the sheet into a DataFrame
        df = excel_data.parse(sheet)
        
        pillar_name = pillar_names[i]
        
        # Initialize an empty list to store the recommendations
        # Iterate through each row in the DataFrame to create the recommendation structure
        for _, row in df.iterrows():
            if isinstance(row["Code"], str):
                recommendation = {
                    #"id": row["Code"],  # "Code" column for the ID
                    "missions": row["Mission_associated"].split(", ") if pd.notna(row["Mission_associated"]) else None,  # Split by comma if not NaN
                    "pillar": pillar_name,
                    "com_b": row["COM-B_component"].strip() if pd.notna(row["COM-B_component"]) else None,  # "COM-B_component" column for COM-B
                    "intervention_types": [t.strip() for t in row["BC_Intervention_Type"].split(", ")] if pd.notna(row["BC_Intervention_Type"]) else []
                }
                recommendations[row["Code"]] = recommendation
        
    for rec_id, rec in recommendations.items():
        for mission_id in rec["missions"]:
            if mission_id not in missions_with_rec_ids:
                missions_with_rec_ids.add(mission_id)
                missions[mission_id] = {"pillar": rec["pillar"]}
                    

    # Convert the result dictionary to a JSON string
    json_data = json.dumps(recommendations, indent=4)
    
    # Optionally, save the JSON data to a file
    with open("utils/contents/recommendations.json", "w") as json_file:
        json_file.write(json_data)
    
    #missions.sort(key=lambda x: (x['id'][:2], int(x['id'][2:])))

    return 

# Example usage:
file_path = "intervention_contents/Recommendations_finished.xlsx"
sheet_names = ["Smoking_finished", "Alcohol_finished", "Nutrition_finished", "Physical_activity_finished",  "Emotional_Wellbeing_finished"]  # Replace with actual sheet names
pillar_names = ["Smoking", "Alcohol", "Nutrition", "Physical Activity",  "Emotional Wellbeing"]  
excel_to_json(file_path, sheet_names, pillar_names)
```

```python
import pandas as pd
import json
import math

def convert_frequency_to_int(str_freq):
    str_to_int = {
        "Once a week": 1,
        "Twice a week" : 2,
        "Three times a week": 3,
        "Four times a week": 4,
        "Five times a week": 5,
        "Six times a week": 7,
        "Daily": 7
    }
    return str_to_int[str_freq]

# Function to process the excel file and generate the JSON structure
def resources_to_json(file_path, sheet_names, pillar_names):

    # Read the excel file
    excel_data = pd.ExcelFile(file_path)

    recommendations = {}

    # Process each sheet
    for i, sheet in enumerate(sheet_names):
        # Read the sheet into a DataFrame
        df = excel_data.parse(sheet)
        
        pillar_name = pillar_names[i]
        
        # Initialize an empty list to store the recommendations
        # Iterate through each row in the DataFrame to create the recommendation structure
        for _, row in df.iterrows():
            if isinstance(row["Code"], str):
                recommendation = {
                    #"id": row["Code"],  # "Code" column for the ID
                    "frequency": convert_frequency_to_int(row["Frequency"]),
                    "pillar": pillar_name.lower()
                }
                recommendations[row["Code"]] = recommendation
        
        
    # Convert the result dictionary to a JSON string
    json_data = json.dumps(recommendations, indent=4)
    
    # Optionally, save the JSON data to a file
    with open("utils/contents/missions.json", "w") as json_file:
        json_file.write(json_data)

    return 

# Example usage:
file_path = "intervention_contents/Missions_finished.xlsx"
sheet_names = ["Smoking_v2_finished", "Alcohol_v2_finished", "Nutrition_finished", "Physical_activity_v2_finished", "Emotional_Wellbeing_finished"]  # Replace with actual sheet names
pillar_names = ["Smoking", "Alcohol", "Nutrition", "Physical Activity", "Emotional Wellbeing"]
resources_to_json(file_path, sheet_names, pillar_names)
```
